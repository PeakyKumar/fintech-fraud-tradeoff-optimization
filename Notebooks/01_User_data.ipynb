```python
import numpy as np
import pandas as pd

np.random.seed(42)

n_users = 50000

user_ids = np.arange(1, n_users + 1)

segments = np.random.choice(
    ['low', 'medium', 'high'],
    size=n_users,
    p=[0.5, 0.35, 0.15]
)

avg_spend = []
churn_prob = []

for seg in segments:
    if seg == 'low':
        avg_spend.append(np.random.uniform(100, 300))
        churn_prob.append(np.random.uniform(0.04, 0.06))
    elif seg == 'medium':
        avg_spend.append(np.random.uniform(300, 800))
        churn_prob.append(np.random.uniform(0.025, 0.035))
    else:
        avg_spend.append(np.random.uniform(800, 3000))
        churn_prob.append(np.random.uniform(0.01, 0.02))

signup_dates = pd.to_datetime(
    np.random.choice(
        pd.date_range("2023-01-01", "2024-12-31"),
        size=n_users
    )
)

tenure_days = (pd.Timestamp("2025-01-01") - signup_dates).days

users = pd.DataFrame({
    "user_id": user_ids,
    "signup_date": signup_dates,
    "user_segment": segments,
    "avg_monthly_spend": avg_spend,
    "base_churn_prob": churn_prob,
    "tenure_days": tenure_days
})

users["lifetime_value"] = (
    users["avg_monthly_spend"] * 
    (1 / users["base_churn_prob"])
)

users.head()
```
